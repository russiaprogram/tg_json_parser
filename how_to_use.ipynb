```python
from tg_json_parser import TgJsonParser
```

### 2. Getting the telegram data

1. Open the channel you want to download, clock on three dots in right top korner of the chat
2. Choose export chat history
3. Change the format from HTML to machine-readable JSON

### 3. Parcing data

```python
# (replace sample data with your own file)
chat_path = 'data/raw/Sample_ChatExport/result.json'
```

```python
# Load the chat data from the JSON file
tg_parcer = TgJsonParser()
tg_parcer.load_json(chat_path)
```

```python
# Parcing data
# (this will take some time depending on the size of the chat)
tg_parcer.extract_messages()
tg_parcer.name
```

```python
# Save the parsed data to a CSV
tg_parcer.save_chat(output_path='data/processed/Sample_Chat')
```

```python
import pandas as pd

pd.DataFrame(tg_parcer.member_actions)
```

```python
# Save the parsed data and mamber_actions to a CSV
tg_parcer.save_chat(output_path='data/processed/Sample_Chat', save_actions=True)
```
